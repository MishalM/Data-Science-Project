```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes_model.csv')
data.head()
```

## Create a Count plot of Outcome

```python
sns.countplot(data['Outcome'], data = data)
```

## Analyse the effect of age on Outcome using Boxplot

```python
sns.boxplot(y= data['Age'], x = data['Outcome'])
```

## Check for the chances of getting diabetes if it is common in family, using the Diabetes Pedigree Function

```python
sns.violinplot(data['Outcome'], data['DiabetesPedigreeFunction'])
```

## Check for the relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
sns.scatterplot(x = data['Insulin'], y = data['Glucose'], data = data, hue = data['Outcome'])
```

## Create a heatmap on Correlation of features in the data.

```python
sns.heatmap(data = data.corr(), annot = True)
```
